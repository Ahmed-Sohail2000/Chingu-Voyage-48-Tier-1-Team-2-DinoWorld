### 1. Load the Json Data

Working with Json Data - https://reintech.io/blog/how-to-work-with-json-data-in-python

````python
# import json, requests, pandas, numpy, matplotlib
import json, requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# setup the json url
dino_url = 'https://raw.githubusercontent.com/chingu-voyages/voyage-project-tier1-dinosaurs/main/assets/dinosaurs.json'
response = requests.get(dino_url)

# open the file
dino_data = json.loads(response.text)

# create a dataframe
dino_data = pd.DataFrame(dino_data)
dino_data
````

````python
# number of dinosaurs eating the different types of diet
from matplotlib import pyplot as plt
import seaborn as sns
dino_data.groupby('diet').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.xlabel('Number of Dinosuars');
plt.title('Total Number of Dinosaurs vs Diet Types');
````

````python
fig, ax = plt.subplots(figsize = (5, 5))

# count the number of diet types
diet_counts = dino_data['diet'].value_counts()

# setup pie chart
ax.pie(diet_counts, labels = diet_counts.index, autopct = '%1.1f%%', shadow = True, explode = [0, 0.1, 0], startangle = 0,
       colors = ['#4392F1', '#F86624', '#615D6C'], pctdistance = 1.17, labeldistance = 0.4);
ax.set_title('Distribution of diet types among dinosaurs');

# export the pie chart as a SVG format
plt.savefig('dino_diet_piechart.svg')
````

````python
# lets plot the dinosaurs found in different countries on a map using plotly
import plotly.express as px
import pandas as pd
!pip install geopy
from geopy.geocoders import Nominatim
import plotly.io as pio
# import json, requests, pandas, numpy, matplotlib
import json, requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# setup the json url
dino_url = 'https://raw.githubusercontent.com/chingu-voyages/voyage-project-tier1-dinosaurs/main/assets/dinosaurs.json'
response = requests.get(dino_url)

# open the file
dino_data = json.loads(response.text)

# create a dataframe
dino_data = pd.DataFrame(dino_data)

# create a function to get the latitude and longitude of the country
def get_lat_lon(country):
  geolocator = Nominatim(user_agent = 'dino_map')
  location = geolocator.geocode(country)

  if location:
    return location.latitude, location.longitude

  else:
    return None, None

# get latitude and longitude for each country and add it into the dino data
dino_data['Latitude'], dino_data['Longitude'] = zip(*dino_data['foundIn'].apply(get_lat_lon))

# remove the rows with missing coordinates
dino_data = dino_data.dropna()

# create the scatter map. we will choose the foundin column from the dino data
fig = px.scatter_geo(dino_data['foundIn'], lat = dino_data['Latitude'], lon = dino_data['Longitude'], color = 'foundIn', hover_name = 'foundIn', title = 'Dinosaurs found around the world', size_max = 5)

# change the background color and make the title bold
fig.update_layout(plot_bgcolor = 'gray')
fig.update_layout(title = {'text' :'Dinosaurs found around the world',
                           'font' : {'size' : 20, 'family' : 'Arial', 'color' : 'black'},
                           'x': 0.5, 'y': 0.95, 'xanchor': 'center', 'yanchor': 'top'})

fig.show()

# export it as a html file
pio.write_html(fig, file = 'dinosaurs around the world.html', auto_open = True)
````

````python
dino_data['length'] = dino_data['length'].replace('N/A', dino_data['length'].value_counts().mean())

# plot the length of the dinosaurs by the type of species
fig, ax = plt.subplots(figsize = (15, 15));

ax.barh(dino_data['typeSpecies'], dino_data['length'], color = plt.cm.tab10.colors);
ax.tick_params(axis = 'y', rotation = 0, size = 10, which = 'major', pad = 15);
ax.set_xlabel('Length of Dinosaurs', size = 15, fontweight = 'bold');
ax.set_ylabel('Types of Dinosaurs', size = 15, fontweight = 'bold');
ax.set_title('Lengths of Dinosaurs Amongst Different Species', size = 15, fontweight = 'bold');

# save the bar plot as a svg type
plt.savefig('dinosaurlength.svg', format = 'svg')
````
